# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def g(x):
  return x

def h(x):
  return -3**(-x)

# Funci칩n cuadr치tica.
def f(x):
  return x - 3**(-x)
```

```python
f(-5), f(1)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-5, 5,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(15,15)) #Tama침o de la figura
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.

plt.grid(color="gray") #Color de la malla
plt.title("Gr치fica de $f(x)=x - 3^{-x}$") #Titulo del grafo
plt.xlabel("Eje X") #Rotulo del eje x
plt.ylabel("Eje Y") #Rotulo del eje y


#plt.figtext(1,1, "g(x) (trazo naranja), h(x) (trazo verde) y f(x) (trazo verde)", ha="center", fontsize="12", color="blue")


plt.show() #Muestra el grafo
```

### Visualizaci칩n gr치fica:
- La gr치fica de $ g(x) $ es una l칤nea recta con pendiente 1.
- La gr치fica de $ h(x) $ es una curva exponencial que decrece a medida que $ x $ aumenta.
- La gr치fica de $ f(x) = x - 3^{-x} $ comienza muy abajo cuando $ x $ es negativo (debido a que $ h(x) $ crece exponencialmente), pasa por $ f(0) = -1 $, y luego se aproxima a la l칤nea recta $ g(x) $ conforme $ x $ aumenta (ya que $ h(x) $ tiende a cero).

Este comportamiento es lo que da origen a la gr치fica de $ f(x) $, que es una combinaci칩n de la recta $ g(x) $ y la curva decreciente $ h(x) $.
