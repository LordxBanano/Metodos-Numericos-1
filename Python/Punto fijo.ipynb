```python
# Código creado por Joshua Mijangos López, 28/09/24
# Este código está hecho para que las funciones del método y sus despejes se modifiquen desde aquí
# Este código no acepta de datos de ingreso desde consola y debe ejecutarse desde Google Colab debido a las líneas de LaTex

from math import *
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex # Esta librería servirá para imprimir en formato LaTex

# Definimos la función f(x) y g(x) para el método del punto fijo
def f(x):
    return x**3 - x - 1  # Esta es la ecuación f(x)

# g(x) = x es la reescritura de f(x) = 0 para obtener una función g(x) que nos permita aplicar el método del punto fijo
def g(x):
    return (x+1)**(1/3)  # Esta es una posible forma de g(x) basada en la ecuación f(x)

# Definimos las funciones en formato LaTeX
latex_f = r"f(x) = x^3 - x - 1"
latex_g = r"x = g(x) = \sqrt[3]{x+1}"

# Imprimos las funciones en formato LaTeX
print("Sean las funciones:\n")
display(Math(latex_f))
display(Math(latex_g))
```

```python
# Ingreso datos de entrada

x0 = 1  # Aproximación inicial (escogida entre el intervalo [1, 2], ya que aquí se encuentra una raíz de mi función f(x), debido al teorema del valor intermedio)

# Tolerancia y número máximo de iteraciones
tol = 10**(-5) # Se debe de editar el número dentro del paréntesis si se busca una solución a f(x) más precisa
nmax = 100
error = 100
niter = 0

print("# de Iteración\t\t    x_n \t\t  g(x_n) \t\t   Error")

# Método del Punto Fijo
while error > tol and niter < nmax:
    x1 = g(x0)

    # En la primera iteración no se imprime el error
    if niter == 0:
        print("\t{0} \t\t {1:6.6f} \t\t {2:6.6f} \t\t {3}".format(niter+1, x0, x1, "   N/A"))
    else:
        error = abs(x1 - x0)
        print("\t{0} \t\t {1:6.6f} \t\t {2:6.6f} \t\t {3:6.6f}".format(niter+1, x0, x1, error))

    # Actualizar x0 para la siguiente iteración
    x0 = x1
    niter += 1

# Imprimir resultado final
print("\nSe realizaron {0} iteraciones.\n".format(niter))
print("La raíz aproximada de la función f(x) es {0:6.6f} con un error de {1:6.5f}".format(x1, tol))
# El parámetro {1:6.zf} debe de modificarse si lo que se busca es disminuir el error, lo dejé preestablecido para z=5 (Es decir, un error de 10^-5)
```
