```python
# Código creado por Joshua Mijangos López, 01/10/24
# Este código está hecho para que las funciones a las que se le aplicará el método de Newton se modifiquen desde aquí
# Este código no acepta datos de ingreso desde consola y se debe ejecutar en Google Colab debido a las líneas de LaTex

from math import *
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex # Para imprimir en formato LaTex
import sympy as sp  # Para cálculo simbólico de la derivada

# Definimos la variable simbólica y la función f(x) para el método de Newton
x_sym = sp.Symbol('x')
f_sym = x_sym**3 - x_sym - 1  # Esta es la ecuación f(x)
df_sym = sp.diff(f_sym, x_sym)  # Calculamos automáticamente la derivada f'(x)

# Convertimos las funciones simbólicas a funciones numéricas para evaluar
f = sp.lambdify(x_sym, f_sym)    # f(x)
df = sp.lambdify(x_sym, df_sym)  # f'(x)

# Definimos las funciones en formato LaTeX
latex_f = r"f(x) = x^3 - x - 1"
latex_df = r"f'(x) = 3x^2 - 1"

# Imprimimos las funciones en formato LaTeX
print("Sean las funciones:\n")
display(Math(latex_f))
display(Math(latex_df))
```

```python
# Ingreso de datos de entrada
x0 = 1  # Aproximación inicial (escogida entre el intervalo [1, 2], ya que aquí se encuentra una raíz de la función f(x))

# Tolerancia y número máximo de iteraciones
tol = 10**(-5)  # Se debe editar el número dentro del paréntesis si se busca una solución más precisa de f(x)
nmax = 100
error = 100
niter = 0

print("# de Iteración\t\t    x_n \t\t   f(x_n) \t\t  Error")

# Método de Newton
while error > tol and niter < nmax:
    fx0 = f(x0)
    dfx0 = df(x0)

    if dfx0 == 0:
        print("La derivada es cero, no se puede aplicar el método de Newton.") # El método se detiene porque no podemos dividir entre cero en la fórmula de Newton.
        break

    x1 = x0 - fx0 / dfx0 # Calcula el siguiente valor x1 usando la fórmula del método de Newton

    # En la primera iteración no se imprime el error
    if niter == 0:
        print("\t{0} \t\t {1:6.6f} \t\t {2:6.6f} \t\t {3}".format(niter+1, x0, fx0, "   N/A"))
    else:
        error = abs(x1 - x0)
        print("\t{0} \t\t {1:6.6f} \t\t {2:6.6f} \t\t {3:6.6f}".format(niter+1, x0, fx0, error))

    # Actualizar x0 para la siguiente iteración
    x0 = x1
    niter += 1

# Imprimir resultado final
if error <= tol:
    print("\nSe realizaron {0} iteraciones.\n".format(niter))
    print("La raíz aproximada de la función f(x) es {0:6.6f} con un error de {1:6.5f}".format(x1, tol))
    # El parámetro {1:6.zf} debe de modificarse si lo que se busca es disminuir el error, lo dejé preestablecido para z=5 (Es decir, un error de 10^-5)
else:
    print("El método no convergió en el número máximo de iteraciones.")
```
